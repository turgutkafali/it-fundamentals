```python
#floatfan sonra kaç hane
text = "{:.2f}, {:.3f}, {:.4f}".format(3.1463, 5.3676, 7.324567)
print(text)
```

```python
#toplam string sayısı
text = "{:.2s} - {:.3s} - {:.4s}".format("3.1463", "5.3676", "7.324567")
print(text)
```

```python
#20 karakterlik bir alan içinde sağa/sola(>/<) yasla
text = "{:<20}".format("Clarusway")
print(text)
print(len(text))
```

```python
#ortalama ^ 
text = "formatlama l{:>10} formatss".format("test")
text
```

```python
# baştan 5 karakterini alıyor
"{:.5}".format("hippopotamus")
```

```python
#toplam 10 karakterlik yere 3 tane al
"{:10.3}".format("hippopotamus")
```

```python
#toplam 10 karakterlik bir alan al sağa yasla 3 tane harf al
"{:>10.3}".format("hippopotamus")
```
